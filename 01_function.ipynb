Basics of function

x -> function -> y

y = f(x) = x^2

y is function of x

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def square(x):
    return np.square(x)
```

Single input - Single output

```python
x = np.arange(-10, 11)
plt.plot(x, square(x))
plt.show()
```

Multiple input - Single Output

```python
# f(x, y) = yx^2
def func1(x, y):
    z = []
    for i in x:
        k = []
        for j in y:
            k.append(j*np.square(i)) 
        z.append(k)
    return np.array(z)
```

```python
x = np.arange(-10, 11)
y = np.arange(-10, 11)

z = func1(x, y)
```

```python
ax = plt.axes(projection='3d')
for i in range(21):
    for j in range(21):
        ax.scatter3D(x[i], y[j], func1(x, y)[i][j])
```
